<font size=5>t-test used to test the correctness of a continous sample

<font size=5>t-test = X - mean/Sx where Sx=s/squarerrotof n  

```python
population = [10,23,45,56,78,23,42,51,78,94,56,72,35,75,120,45]
```

```python
import numpy as np
mean = np.mean(population)
print(mean)
```

```python
sample = np.random.choice(population,10)
print(sample)
```

```python
np.mean(sample)
```

```python
from scipy.stats import ttest_1samp
ttest,p_value = ttest_1samp(sample,56)
```

```python
print("tetst value is",ttest)
print("p_value is",p_value)
```

```python
if p_value < 0.05:
    print("we are rejecting null hypothesis")
else:
    print("we are accepting null hypothesis")
```

```python
import pandas as pd
import scipy.stats as st
```

```python
age_population = st.poisson.rvs(loc=18,mu=35,size=1500)
age_sample = st.poisson.rvs(loc=18,mu=30,size=60)
```

```python
np.mean(age_population)
```

```python
np.mean(age_sample)
```

```python
ttest,pvalue = st.ttest_1samp(a=age_sample,popmean=np.mean(age_population))
```

```python
print(pvalue)
```

```python
if pvalue < 0.05:
    print("we are rejecting null hypothesis")
else:
    print("we are accepting null hypothesis")
```

<font size=5>t-test for multiple samples

```python
import numpy as np
import scipy.stats as st
```

```python
height_population = st.poisson.rvs(loc=18,mu=35,size=1500)
height_sample1 = st.poisson.rvs(loc=18,mu=30,size=60)
height_sample2 = st.poisson.rvs(loc=18,mu=25,size=60)
```

```python
print("mean of population is",np.mean(height_population))
print("mean of sample1 is",np.mean(height_sample1))
print("mean if sample2 is",np.mean(height_sample2))
```

```python
test,pvalue = st.ttest_ind(height_sample1,height_sample2)
```

```python
print(pvalue)
```

```python
if pvalue < 0.05:
    print("we are rejecting null hypothesis")
else:
    print("we are accepting null hypothesis")
```

<font size=5>paired t-test:it is used inorder to compare samples of same group

```python
weight1 = [25,30,34,54,27,39,45,54,38]
weight2 = weight1+st.norm.rvs(scale=5,loc=-1.25,size=9)
```

```python
print(weight1)
print(weight2)
```

```python
ttest,p_value = st.ttest_rel(weight1,weight2)
```

```python
print(p_value)
```

```python
if p_value < 0.05:
    print("we are rejecting null hypothesis")
else:
    print("we are accepting null hypothesis")
```

```python
import pandas as pd
df1 = pd.read_csv('mobile.csv')
df1.head()
```

```python
print(independent.shape)
print(dependent.shape)
```

```python
independent.isnull().sum()
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
best_feature = SelectKBest(score_func=chi2,k=20)
best_feature.fit(independent,dependent)
```

```python
score = best_feature.scores_
print(score)
```

```python
df2 = pd.DataFrame(score)
df2['features'] = independent.columns
```

```python
df2.columns=['scores','features']
```

```python
df2
```

```python
a = df2.nlargest(10,'scores')
a.head()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=[10,10])
plt.xticks(rotation=40)
plt.barh(a["features"],a["scores"])
```

```python
threshold=0.5
```

```python
# find and remove correlated features
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```
