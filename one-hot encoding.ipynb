```python
df1 = pd.read_csv("train.csv",usecols=['X1','X2','X3','X4','X5'])
df1.head()
```

```python
top10_x1 = [features for features in df1.X1.value_counts().sort_values(ascending=False).head(10).index]
print(top10_x1)
```

```python
top10_x2 = [features for features in df1.X2.value_counts().sort_values(ascending=False).head(10).index]
print(top10_x2)
```

```python
top10_x3 = [features for features in df1.X3.value_counts().sort_values(ascending=False).head(10).index]
print(top10_x3)
```

```python
top10_x4 = [features for features in df1.X4.value_counts().sort_values(ascending=False).head(10).index]
print(top10_x4)
```

```python
top10_x5 = [features for features in df1.X5.value_counts().sort_values(ascending=False).head(10).index]
print(top10_x5)
```

```python
def one_hot(df1,top10,column):
    for x in top10:
        df1[x] = np.where(df1[column]==x,1,0)
        
        
one_hot(df1,top10_x1,"X1")
```

```python
one_hot(df1,top10_x2,"X2")
```

```python
one_hot(df1,top10_x3,"X3")
```
